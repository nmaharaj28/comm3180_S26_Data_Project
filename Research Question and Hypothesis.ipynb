# Main Research Question and Hypothesis

## Main Research Question

What are the demographic profiles (e.g., gender, race/ethnicity, age, and primary discipline) of winners across the Emmy, Grammy, Oscar, and Tony Awards, and how do these patterns compare across awards while reflecting broader structural and industry-level factors that shape differences in representation?

## Hypothesis

Null hypothesis: There are no significant demographic differences between winners of the Emmy, Grammy, Oscar, and Tony Awards.

Alternative hypothesis: Winners of the Emmy, Grammy, Oscar, and Tony Awards are disproportionately white and male.
