# Project Plan -- Steps in the Project

### Data Collection and Sorting

We will start by downloading and loading the four primary datasets of the Emmy, Grammy, Oscar, and Tony winners. We will supplement Emmy and Oscar files with official database records for 2015–2026

After downloading the data sets we will, hand-compile the EGOT master spreadsheet (n ≈ 28) from Wikipedia, Britannica, and Billboard, with columns for: name, year_E, year_G, year_O, year_T, year_egot_completed, age_at_completion, nationality, primary_discipline, gender, race_ethnicity. 

We will then code demographic variables for each winner: gender (from public self-identification), race/ethnicity (researcher-coded from Wikipedia and IMDb biographies into six broad categories), and nationality (country of birth / primary citizenship at time of first win); document all coding decisions in a codebook. We might explore the possibility of using Natural Language Processing and AI to streamline demographic identification process and identify nation/city of origin.

### Pre-Award Analysis

For each of the four awards we will separately, calculate frequency distributions of winners by gender, race/ethnicity, nationality, and primary discipline. We will visualize this as percentages of total winners in each category for comparability across awards with different total winner counts.

For our analysis, we will focus on performing categories for the Emmys, Oscars, and Tonys. For example, we will be focusing on the “Big 4” for the Grammys (Album of the Year, Song of the Year, Record of the Year, and Best New Artist)/ 

Then, we will group winners by decade (1930s–2020s) and calculate the proportion of women and people of color per decade for each award. We plan on visualizing this as line or area charts — the primary visualization for tracking whether and when each award became more diverse over time. 

### Cross-Award Comparison

For cross-award comparsion, we will plot side-by-side grouped bar charts displaying gender and racial composition of winners for all four awards simultaneously. This allows for direct visual comparison of which award looks most and least diverse on each demographic dimension.

Then, we will run chi-square tests of independence to assess whether differences in demographic composition across awards are statistically significant.

For awards with documented institutional reform — Oscar (2016 AMPAS diversity initiative), Grammy (2020 Recording Academy task force) — we will split the dataset at the reform year and run chi-square comparisons of pre- vs. post-reform demographic distributions within that award.

We will then examine category count as an explanatory variable: test whether the Grammy's broader category architecture (95+) produces more demographic variety than the Oscar and Tony's narrower structures (~24–26 categories), and whether diversity gains at the Grammys are concentrated in newly added genre categories vs. long-standing ones.

### EGOT Combined analysis

We are also interested in looking at EGOT winners, individuals who have won all four awards. 

There are approximately 28 EGOT winners total — 22 who won all four awards competitively, and 6 who completed the set with at least one honorary or special award. We will treat these two groups separately throughout.

Note that every non-American EGOT winner to date is British or Western European — no winner has come from Latin America, Asia, Africa, or the Global South despite those regions producing internationally recognized artists across all four relevant industries.

We will compare the demographic makeup of EGOT winners against the broader winner pools of each individual award.

We will then compare the order in which each winner received their four awards (e.g., Oscar first, then Tony, then Grammy, then Emmy) and visualize all pathways.

Finally, we will identify the most common entry point — which award do most EGOT winners collect first — and whether that first award tends to predict how quickly the remaining three follow.
