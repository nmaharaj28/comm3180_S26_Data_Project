# Project Plan for "Does Diversity Win? Oscar Nominees, Award Winners, and the Box Office Before and After #OscarsSoWhite

### Phase 1: Oscar Nominations & Winners DataData source: Multiple Kaggle datasets + GitHub curated set

Datasets: 
* Nominations + winners with IMDb IDs: github.com/DLu/oscar_data. This is the most carefully maintained dataset; uses IMDb unique identifiers to disambiguate people with the same name across years
* Demographics of race, gender, religion: kaggle.com/datasets/fmejia21/demographics-of-academy-awards-oscars-winners. Covers winners since 1928 with demographic fields already attached
* Full nominations 1927–2025: kaggle.com/datasets/unanimad/the-oscar-award — broadest year coverage, use to fill gaps
* Cross-check source: UCLA Diversity & Oscars project dataset (see docs/codebooks/ for their R cleaning methodology.

While cleaning the dataset, add a post_campaign column: 0 for years ≤ 2015, 1 for 2016 onward

Focus categories for v1: Best Picture, Best Director, Best Actor, Best Actress, Best Supporting Actor, Best Supporting Actress, Best Original Screenplay, Best Adapted Screenplay. Other categories can be added in v2.

Can answer the following questions with the data:

 - Did the share of non-white nominees in acting categories increase after 2016, and what was the trend?
- Is there a gap between nomination rates and win rates for non-white nominees — i.e., do nominees of color get nominated but not win at disproportionate rates?
- In what categorires did diversity gains first appear in (if they appeared at all).
- Did gender diversity follow the same trajectory as racial diversity after 2016, or did the two move independently?

### Phase 2: What is winning worth?

Datasets: 
* kaggle.com/datasets/parthdande/movies-box-office-collection-data-2000-2024 — domestic and global gross
* the-numbers.com/movie/budgets — production budgets, needed to compute ROI

Can answer the following questions with the data

* Do Oscar-winning films earn more at the box office than films that were nominated but did not win?
* Did the box office value of winning an Oscar change after 2016?
* Do Oscar-winning films generate a better return on investment than nominated-but-losing films, controlling for production budget?
* Do Oscar-winning films with a non-white lead actor earn more or less at the box office than Oscar-winning films with a white lead, controlling for production budget and genre?
* Do films with a non-white Best Director nominee earn differently than films with a white Best Director nominee, and did that change after 2016?
* Are diverse-led Oscar winners given smaller production budgets than non-diverse Oscar winners, and has that budget gap narrowed post-2016?
* Among Best Picture winners specifically, does diversity of the cast correlate with global box office performance?
