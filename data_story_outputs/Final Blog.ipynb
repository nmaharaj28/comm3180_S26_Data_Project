## Introduction and Research Questions

The movie and entertainment industry is one of the most public facing industries in the world. From actors and actresses to producers and directors, these individuals come together to produce content and media for the audiences across the globe to consume. However, with an industry creating content for a large and diverse audience, social media has been quick to point out the lack of diversity in these films. 

More specifically, the Oscars—the most esteemed and established award show for the film industry—was under heavy criticism in the mid-2010s for its lack of racial diversity in its nominations and awards. The criticism, housed under the trending #OscarsSoWhite Twitter movement sparked global conversations about racial diversity in film. 

#OscarsSoWhite was a trending hashtag in 2016 that called attention to the fact that the majority of Oscar nominees and winners were overwhelmingly white.

With the public criticism of the lack of racial diversity in the Oscars making headlines, we decided to analyze whether the #OscarsSoWhite movement produced a lasting, quantifiable shift in Oscar nomination diversity, box office performance of films with diverse leads, and cultural conversation. 

Using data on the racial demographics of Oscar nominees and winners from the last 98 years, historical box office data, and a sample size of tweets containing #OscarsSoWhite, multiple research questions emerged.

Notably, how have people of color nomination and win rates changed over time? Did POC nomination rates in acting and directing categories increase significantly after 2016? Do films with POC leads earn differently at the box office than white-lead films, and did that relationship change post-2016? How, if at all, has #OscarsSoWhite influenced the cultural discourse surrounding Academy Award diversity?

### Hypothesis

We hypothesize that POC nomination rates increased significantly post-2016, driven by the Oscar’s membership expansion rather than gradual organic change. Additionally, pre-2016, POC-lead nominees earned less domestically than white-lead nominees; post-2016, that gap narrowed as the Oscar nominated higher-profile diverse films.

### Who Gets Nominated and Who Wins?
The Academy Awards have existed since 1929, but for most of that history the people in contention for them have looked remarkably similar: predominantly male, overwhelmingly white. This section traces the demographics across 98 ceremonies through the lens of gender, and race to establish what was the catalyst for #OscarsSoWhite, and how demographics have changed since. 

*Gender: Slow Progress, a Persistent Ceiling*

The Oscars have always been a male-dominated institution, and the raw count of nominations by year tells that story visually.

Across all 11,176 nominee records spanning 98 years, women make up just 21.1% of nominees overall — and 19.7% of winners. Those two numbers being nearly identical is itself a finding: once nominated, women win at roughly the same rate as men (21.4% vs. 22.8%), meaning the gender gap lives almost entirely at the nomination stage, not in the final vote.

The decade-by-decade proportion chart reveals the pace of that progress — and its limits. Female representation among nominees grew from a low of 9.5% in the 1940s to 28.8% in the 2000s and 31.0% in the 2010s, then essentially plateaued at 30.6% in the 2020s. The #OscarsSoWhite era did not accelerate female nomination rates the way it did for racial diversity. Gender equity at the Oscars has been a story of gradual, incomplete progress.

Male and female nominees have converted nominations to wins at nearly identical rates across every decade. The nomination bottleneck is where the gender disparity is.

*Where Women Are Most Absent: Key Non-Acting Categories*

The aggregate female nomination rate of ~21% is already low, but it is inflated by acting categories (which are segregated by gender), guaranteeing 40–50% female nominees in any year. When we look at the categories that carry the most institutional prestige such as Directing, Cinematography, Film Editing, Best Picture, and both Writing categories, the gender gap increases.

The Directing category is the most striking case. Over 98 ceremonies, only 2.9% of Directing nominees have been women. That figure was 0% from the 1920s through the 1940s, barely reached 4% at its pre-2020s peak, and jumped to 17.1% in the 2020s. While this is low, it is the sharpest decade-over-decade increase in the category's history. Cinematography and Film Editing show similar patterns: near-zero female representation for most of the Academy's history, with only tentative movement in recent decades. Best Picture and the Writing categories show slightly better but still deeply unequal numbers.

The post-2016 membership expansion that diversified the Academy racially appears to have had some spillover effect on gender representation in technical and creative categories, but far less dramatically than its effect on racial diversity.

*Race*

If the gender story is one of slow, incomplete progress, the racial story is one of near-total exclusion followed by a sudden, statistically dramatic shift.

For most of the Academy's history, the non-white bands are barely distinguishable from zero. Across all 98 ceremonies, people of color constitute just 6.0% of all nominees — and that number is pulled upward almost entirely by recent years.

POC nominees accounted for less than 5% of nominations in every decade through the 2000s. Then: 3.7% pre-2016, jumping to 24.3% post-2016. The winning rate followed an even more dramatic arc — POC winners went from 3.8% of all winners before 2016 to 37.4% afterward. In the 2020s alone, 27.0% of nominees have been people of color.

*Where Racial Diversity Has and Hasn't Changed*

In acting and directing (the categories most publicly associated with #OscarsSoWhite) the pre/post contrast is especially sharp. POC nominees in these categories rose from 10.5% pre-2016 to 25.2% post-2016 (χ²(1) = 23.47, p < 0.000001). The year-by-year panel shows just how little pre-reform movement there was: the line bumps along near 10% for 15 years, then steps decisively upward after 2017 and holds there.

The win rate by race chart surfaces a counterintuitive finding: once nominated, people of color are not disadvantaged at the final vote. In 2020, Black nominees won at a 26.6% rate, Hispanic nominees at 24.5%, and Asian nominees at 22.5% — all at or above the white nominee win rate of 22.4%. As with gender, the racial inequity at the Oscars is concentrated at the nomination gate, not in the final voting room.


The category grid reveals where change has and hasn't penetrated. Acting categories show the clearest post-2016 diversification, which is consistent with where public pressure was most concentrated. Best Picture and Directing show meaningful but smaller movement. Cinematography, Film Editing, and Writing remain overwhelmingly white even in the most recent decade, suggesting that the #OscarsSoWhite intervention has been substantially more effective in front-of-camera categories than in the technical and creative crafts that shape films behind the scenes.

*Conclusion*

Across 98 ceremonies, the Academy's demographic record divides into two eras with unusual clarity. Before 2016: a century of near-total white male dominance, with women slowly gaining ground to about 30% of nominations while people of color remained in the low single digits across virtually every category. After 2016: a sharp break in racial diversity, particularly in acting and directing, with POC nominees rising to roughly a quarter of nominations and POC winners reaching over a third—numbers that would have been unimaginable a decade earlier.

The gender story is more complicated. #OscarsSoWhite was explicitly a racial justice campaign, and the data shows its effects were correspondingly concentrated in racial diversity. Female representation has continued its gradual climb, with the 2020s showing meaningful movement in Directing specifically, but no equivalent step change to what happened with race.

Once nominated, women and people of color win at rates comparable to or exceeding white men. The Academy's diversity problem has always been about who gets in the room.

### Box Office Analysis
If diverse-lead films were being systematically overlooked by the Academy, was the industry leaving money on the table, or were the Oscars simply indifferent to films that audiences were already rewarding at the box office?

To answer this, we analyzed domestic box office data for Oscar-nominated films from 2001 to the present, joining them to the nomination and demographics dataset by film title. 

*Who gets nominated and what their films earn*

Looking at all Best Picture nominees from 2001 to the present with box office data (n = 176 films), the overall median domestic gross for POC-lead nominees was $68.0M, compared to $56.8M for white-lead nominees. On the raw numbers, POC-lead films actually out-earned white-lead nominees at the median.

Pre-2016, POC-lead Best Picture nominees earned a median of $107.1M, which is considerably more than white-lead nominees at $71.5M. Films like The Help (2011), 12 Years a Slave (2013), Selma (2014), and Straight Outta Compton (2015) were commercially ambitious releases that crossed over into mainstream audiences. Post-2016, both groups dropped — white-lead nominees fell to $34.6M and POC-lead nominees to $50.5M — and the gap narrowed rather than widened.

The right panel confirms that across all years combined, the distributions overlap substantially, with POC-lead films holding a slightly higher median ($68M vs. $57M). A Mann-Whitney U test finds this overall difference is not statistically significant (U = 3,349, p = 0.774).

*Do Diverse-Lead Films Earn Differently Across All Categories?*

Best Picture is only one category. Extending the analysis to all major Oscar categories reveals whether the race-gross relationship is consistent across the types of recognition being given.
The pattern here is striking and consistent: across every single category, POC-lead nominees outearned white-lead nominees at the median:

The Directing category shows the largest gap ($76.3M vs. $47.8M), reflecting high-profile films with POC directors that also crossed over commercially. The Actress categories show the lowest absolute grosses overall, which is consistent with the well-documented industry pattern that female-led films are greenlit at lower budgets.


*Regression Analysis Shows No Significant Effect*
The OLS regression predicting log domestic gross from lead race, post-2016 era, Best Picture win, and a POC × post-2016 interaction term produced an R² of just 0.061. None of the four predictors reached statistical significance:

Our test of whether POC-lead films performed differently relative to white-lead films after the reform is negative and non-significant (p = 0.234). Thus, we cannot conclude that the Academy's post-2016 diversification produced a detectable shift in the relative commercial performance of diverse-lead films. This null result is itself meaningful: Oscar recognition, at the nominee level, may have weaker commercial influence than commonly assumed.

*Widening Distance From Blockbusters*
Finally, how do Best Picture nominees compare to what audiences were actually paying to see each year?

The divergence is dramatic. In 2001, the gap between the top 10 median ($214M) and Best Picture nominees ($126M) was $88M. By 2015–2016, that gap had exploded to nearly $300M. Post-2016, two forces collided: COVID collapsed theatrical grosses in 2020–2021, and in the recovery years of 2022–2024 the gap reopened to $300–360M as blockbusters rebounded while Oscar nominees became increasingly international, arthouse, and streaming-adjacent, earned just $13–21M at the domestic box office.
The Academy diversified its nominee pool but continued to favor films that are culturally significant and commercially distant from the mainstream. The #OscarsSoWhite campaign changed who gets recognized. It did not change the Academy's preference for prestige over box office.
*Conclusion*
There was no pre-2016 earnings penalty for POC-lead Oscar nominees. If anything, the pre-reform POC nominees were high-profile commercial films that out-earned white-lead nominees at the median ($107M vs. $72M). Post-reform, both groups saw lower median grosses as the Academy's tent expanded to include more international and arthouse films. The regression finds no statistically significant race-era interaction effect.
What the data does show is that the Academy's reform produced a more diverse nominee pool without fundamentally altering the commercial profile of what it chooses to recognize. 

## Analysis of #OscarsSoWhite Tweets

Using a sample of 1% of all tweets containing #OscarsSoWhite, we conducted a sentiment analysis to categorize Tweets for different categories of support or criticism of #OscarsSoWhite. We then also looked at the frequency of #OscarsSoWhite Tweets throughout major years of controversy for Oscar diversity.

*What are attitudes toward #OscarsSoWhite?*

We conducted a sentiment analysis using an LLM to determine the attitudes toward #OscarsSoWhite and determined how people were using the hashtag. We gave the LLM three concepts. Concept 1 representing positive, supportive, and solidarity with #OscarsSoWhite and a general support for an increase in diversity in the Oscars. Concept 2 was a criticism of #OscarsSoWhite including opposition, mocking, or dismissing the #OscarsSoWhite movement. Finally, the last concept is neither, representing neutral news articles using the hashtag, unreadable or spam tweets, and general neutral tweets that did not take a stance on the movement.


The graph above displays the distribution of the fully coded tweet dataset across all three stance categories. The majority of tweets fell into the “neither” category, indicating that many posts were either neutral news reports discussing the movement or tweets that could not be clearly interpreted. The second-largest category was Concept 1, representing support for the #OscarsSoWhite movement and accounting for more than one-third of the dataset. Finally, approximately 1,500 tweets were categorized as Concept 2, representing criticism of the movement.

The chart implies that the #OscarsSoWhite movement was mostly discussed positively on Twitter, with negative sentiment being heavily outnumbered by positive and supportive stances and neutral tweets. 
The large number of “neither” tweets suggests that #OscarsSoWhite generated widespread media coverage and sparked a broader national conversation about diversity in film. The neutral tweets indicate a large number of news platforms reporting on the movement, implying that the movement created large public conversations. 
The presence of a substantial number of critical tweets is particularly notable given that users employing the hashtag might be expected to predominantly support the movement. The large number of Concept 2 tweets demonstrates that the movement was contested. While reading the tweets, some of the criticism against the movement was that #OscarsSoWhite was only advocating for a greater representation of Black actors and producers in the film industry, rather than advocating for other people of color. Notably, criticism included a lack of  advocacy for Asian and Hispanic diversity in the Oscars.

The graph above illustrates the distribution of tweet concepts from 2015 to 2022. Across every year, the “neither” category remained the most common stance, with Concept 1 as second, and Concept 2 as third. This suggests that most of the online conversation consisted of neutral stances. 
*Timeline of #OscarsSoWhite*

The first analysis was an overview graph of the yearly frequency of #OscarsSoWhite from 2015 to 2022.

The graph above demonstrates that the hashtag reached its peak usage in 2016, one year after its creation, suggesting that the movement required time to gain national attention and generate broader public discourse. However, hashtag usage declined sharply in 2017, indicating a reduction in sustained public conversation surrounding diversity within the Oscars. Between 2019 and 2022, usage remained relatively low compared to the heightened activity of 2015 and 2016. Following a yearly overview, we conducted an analysis to determine the frequency of #OscarsSoWhite by day, to gain a better understanding of what time of year users are discussing Oscar’s diversity. 

The graph above indicates that, in 2015 and 2016, the majority of tweets containing #OscarsSoWhite were published in two waves. In 2018 and 2020 for example, we see the same pattern of a high frequency during nominations and then during the Oscar’s ceremony.

This suggests that conversations surrounding diversity in films are the most salient and prominent when the Oscars are in the news and on the top of the audience's minds. There is little advocacy for diversity in elite films when the Oscars are not announcing their nominations or hosting their ceremony.

In 2015, the hashtag was created, it saw peak use in 2016, and a small resurge in 2017 after the La La Land and Moonlight controversy, where the white-led cast of La La Land was wrongly announced as the winner of Best Picture instead of the actual winner, Moonlight, a POC-led cast.

## Synthesis

Together, the demographic, box office, and Twitter analyses demonstrate that #OscarsSoWhite had a large impact and influence in the Oscar’s nomination patterns and shaped public discourse surrounding diversity in films. 

The demographic data shows the evidence of the #OscarsSoWhite movement influenced the nominations of the Oscars as people of color comprised almost a quarter of nominations post-2016. While this is a strong contrast from the gender gap of Oscar nominees, it still represents a positive change toward an increase in diversity. At the same time, the win-rate analysis demonstrated that women and people of color were not significantly disadvantaged once nominated. These findings suggest that #OscarsSoWhite functioned as a pressure campaign that successfully challenged the Academy.

Additionally, the box office data emphasized that POC-led films did not suffer from weaker commercial performance; in many categories, they actually outperformed white-led nominees at the median. The Twitter analysis also demonstrated how #OscarsSoWhite fostered national conversations about diversity in filmmaking. While the hashtag peaked in usage closer to Oscar nominations and ceremonies, the hashtag was able to captivate national attention and spotlight the discrepancy in Oscar nominations. The hashtag has lost salience after 2017, suggesting that conversations about diversity in the Oscars are no longer at the top of viewers' minds. 

Taken together, these three forms of analysis show that, while #OscarsSoWhite can be overlooked as a trendy hashtag, it worked to influence Oscar nominations and led national conversations about diversity in filmmaking.
