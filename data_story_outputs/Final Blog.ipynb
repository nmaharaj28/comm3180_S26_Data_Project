## Introduction and Research Questions

The movie and entertainment industry is one of the most public facing industries in the world. From actors and actresses to producers and directors, these individuals come together to produce content and media for the audiences across the globe to consume. However, with an industry creating content for a large and diverse audience, social media has been quick to point out the lack of diversity in these films. 

More specifically, the Oscars—the most esteemed and established award show for the film industry—was under heavy criticism in the mid-2010s for its lack of racial diversity in its nominations and awards. The criticism, housed under the trending #OscarsSoWhite Twitter movement sparked global conversations about racial diversity in film. 

#OscarsSoWhite was a trending hashtag in 2016 that called attention to the fact that the majority of Oscar nominees and winners were overwhelmingly white.

With the public criticism of the lack of racial diversity in the Oscars making headlines, we decided to analyze whether the #OscarsSoWhite movement produced a lasting, quantifiable shift in Oscar nomination diversity, box office performance of films with diverse leads, and cultural conversation. 

Using data on the racial demographics of Oscar nominees and winners from the last 98 years, historical box office data, and a sample size of tweets containing #OscarsSoWhite, multiple research questions emerged.

Notably, how have people of color nomination and win rates changed over time? Did POC nomination rates in acting and directing categories increase significantly after 2016? Do films with POC leads earn differently at the box office than white-lead films, and did that relationship change post-2016? How, if at all, has #OscarsSoWhite influenced the cultural discourse surrounding Academy Award diversity?

### Hypothesis

We hypothesize that POC nomination rates increased significantly post-2016, driven by the Oscar’s membership expansion rather than gradual organic change. Additionally, pre-2016, POC-lead nominees earned less domestically than white-lead nominees; post-2016, that gap narrowed as the Oscar nominated higher-profile diverse films.

### Who Gets Nominated and Who Wins?
The Academy Awards have existed since 1929, but for most of that history the people in contention for them have looked remarkably similar: predominantly male, overwhelmingly white. This section traces the demographics across 98 ceremonies through the lens of gender, and race to establish what was the catalyst for #OscarsSoWhite, and how demographics have changed since. 

*Gender: Slow Progress, a Persistent Ceiling*

The Oscars have always been a male-dominated institution, and the raw count of nominations by year tells that story visually.

Across all 11,176 nominee records spanning 98 years, women make up just 21.1% of nominees overall — and 19.7% of winners. Those two numbers being nearly identical is itself a finding: once nominated, women win at roughly the same rate as men (21.4% vs. 22.8%), meaning the gender gap lives almost entirely at the nomination stage, not in the final vote.

The decade-by-decade proportion chart reveals the pace of that progress — and its limits. Female representation among nominees grew from a low of 9.5% in the 1940s to 28.8% in the 2000s and 31.0% in the 2010s, then essentially plateaued at 30.6% in the 2020s. The #OscarsSoWhite era did not accelerate female nomination rates the way it did for racial diversity. Gender equity at the Oscars has been a story of gradual, incomplete progress.

Male and female nominees have converted nominations to wins at nearly identical rates across every decade. The nomination bottleneck is where the gender disparity is.

*Where Women Are Most Absent: Key Non-Acting Categories*

The aggregate female nomination rate of ~21% is already low, but it is inflated by acting categories (which are segregated by gender), guaranteeing 40–50% female nominees in any year. When we look at the categories that carry the most institutional prestige such as Directing, Cinematography, Film Editing, Best Picture, and both Writing categories, the gender gap increases.

The Directing category is the most striking case. Over 98 ceremonies, only 2.9% of Directing nominees have been women. That figure was 0% from the 1920s through the 1940s, barely reached 4% at its pre-2020s peak, and jumped to 17.1% in the 2020s. While this is low, it is the sharpest decade-over-decade increase in the category's history. Cinematography and Film Editing show similar patterns: near-zero female representation for most of the Academy's history, with only tentative movement in recent decades. Best Picture and the Writing categories show slightly better but still deeply unequal numbers.

The post-2016 membership expansion that diversified the Academy racially appears to have had some spillover effect on gender representation in technical and creative categories, but far less dramatically than its effect on racial diversity.
